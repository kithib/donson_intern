```python
import pandas as pd
from paddlenlp import Taskflow
```

```python
df_1 = pd.read_excel('predict_test_1.xlsx')
df_2 = pd.read_excel('predict_test_2.xlsx')
df_3 = pd.read_excel('predict_test_3.xlsx')
df_4 = pd.read_excel('predict_test_4.xlsx')
df_1
```

```python
frames = [df_1, df_2,df_3,df_4]
df = pd.concat(frames) 
```

```python
data_columns = ['content_clean' ]
df_data = df[data_columns]
df_data
```

```python
df_data.drop_duplicates(subset=['content_clean'],keep='first',inplace=True)
df_data
```

```python
for i in range(int(df_data.shape[0])):
    df_data.iloc[i,0] = str(df_data.iloc[i,0]).replace('\t','')
    df_data.iloc[i,0] = str(df_data.iloc[i,0]).replace('\n','')
```

```python
df_data = df_data.reset_index(drop = True)
df_data
```

```python
df_data = df_data.assign(predict_label = None, score = None)
cls = Taskflow("text_classification", task_path='checkpoint/export', is_static_model=True)
for i in range(int(df_data.shape[0])):
    df_data.iloc[i,1] = cls(str(df_data.iloc[i,0]))[0]["predictions"][0]["label"]
    df_data.iloc[i,2] = cls(str(df_data.iloc[i,0]))[0]["predictions"][0]["score"]
    print(str(i) + " , " + str(df_data.shape[0]))
```

```python
df_data.to_excel('predict_data.xlsx', index=False, engine='openpyxl')
```
