```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_excel('data_real.xlsx')
sorted_df = df.sort_values(by=['label', 'score'], ascending=[False, False]).reset_index(drop=True)
sorted_df
```

```python
df_985 = sorted_df
for index, row in df_985.iterrows():  
    if row["label"]  == 1 :
       df_985 = df_985.drop(index) 
    else:
        if row["score"]  < 0.985 :
           df_985 = df_985.drop(index)  
df_985  = df_985.reset_index(drop=True)
df_985
```

```python
df_97 = sorted_df
for index, row in df_97.iterrows():  
    if row["label"]  == 1 :
       df_97 = df_97.drop(index) 
    else:
        if row["score"]  < 0.97 :
           df_97 = df_97.drop(index)  
df_97  = df_97.reset_index(drop=True)
df_97
```

```python
df_95 = sorted_df
for index, row in df_95.iterrows():  
    if row["label"]  == 1 :
       df_95 = df_95.drop(index) 
    else:
        if row["score"]  < 0.95 :
           df_95 = df_95.drop(index)  
df_95  = df_95.reset_index(drop=True)
df_95
```

```python
np.random.seed(seed=2)
df_985 = df_985.sample(frac=0.05)
df_97 = df_97.sample(frac=0.05)
```

```python
df_985 = df_985.sort_values(by=['score'], ascending=[False]).reset_index(drop=True)
df_97 = df_97.sort_values(by=['score'], ascending=[False]).reset_index(drop=True)
df_985,df_97
```

```python
df_985.to_excel('data_985.xlsx', index=False, engine='openpyxl')
df_97.to_excel('data_97.xlsx', index=False, engine='openpyxl')
```
