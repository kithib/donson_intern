```python
import json
import re
```

```python
with open('obs_data.json', 'r') as f:
    lines = f.readlines()
data = []
for line in lines:
    try:
        data_for_line = json.loads(line)
        # 如果解析成功，则添加到数据列表中
        data.append(data_for_line)
    except json.JSONDecodeError:
        # 打印出无法解析的行并跳过
        print(f"Skipping invalid JSON line: {line}")
len(data)
```

```python
def clean_text_1013(cls, text):
    text = str(text)
    # 定义正则表达式
    pattern1 = re.compile(r'\[.*?\]//@.*?:')
    pattern2 = re.compile(r'#.*?#')
    pattern3 = re.compile(r'//@.*?:')
    pattern4 = re.compile(r'#.*?\[.*?\]#')
    pattern5 = re.compile(r'\(http://t\.cn/\w+\)')
    pattern6 = re.compile(r'(CR|cr|Cr)[:：]?\s*\w+')
    pattern7 = re.compile(r'@(\w+)')
    pattern8 = re.compile(r'\[[^\]]*的微博视频\]')

    # 清洗文案
    clean_text = re.sub(pattern1, "", text)
    clean_text = re.sub(pattern2, "", clean_text)
    clean_text = re.sub(pattern3, "", clean_text)
    clean_text = re.sub(pattern4, "", clean_text)
    clean_text = re.sub(pattern5, "", clean_text)
    clean_text = re.sub(pattern6, "", clean_text)
    clean_text = re.sub(pattern7, "", clean_text)
    clean_text = re.sub(pattern8, "", clean_text)

    return clean_text
    

def clean_text_1027(text):
    text = str(text)

    # 正则表达式清洗#xxx[xx]#，例如#我的追剧宝藏零食[话题]#
    pattern1 = r"#.+?\[话题\]#"

    # pattern = r"#.+?\[\w+\]#"
    # pattern = r"#.+?\[(话题|超话|超会|粉丝群)\]#"

    # 清洗文案
    _text = re.sub(pattern1, "", text)
    # 正则表达式清洗@xxx，例如@亚米-Yami；#xxx，例如#笔记灵感
    pattern2 = r"#\S+|@\S+"
    # 清洗文案
    _text = re.sub(pattern2, "", _text)
    _text = _text.replace(" ", "")
    _text = _text.replace('"', "")
    _text = _text.replace("\n", "")
    _text = _text.replace("\t", "")
    _text = _text.replace("\\", "")
    _text = re.sub(r'[^A-Za-z\u4e00-\u9fa5]', '', _text)
    # 把所有英文单词转为大写
    _text = _text.upper()

    return _text
```

```python
for i in range(len(data)):
    data[i]['content'] = clean_text_1027(data[i]['content'])
print(len(data))
```

```python
del_list = []
for i in range(len(data)):
    if data[i]['source'] != '1027':
        del_list.append(i)
tem = [item for index, item in enumerate(data) if index not in del_list]
data = tem
print(len(data))
data
```
