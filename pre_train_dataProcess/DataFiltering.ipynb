```python
import pandas as pd
```

```python
df_result = pd.read_excel("whole_result.xlsx")
df_result = df_result.dropna()
df_result = df_result.drop_duplicates(subset='语句内容').reset_index(drop=True)
df_result
```

```python
last_score = df_result.iloc[0,4]
for index, row in df_result.iterrows():  
    if index != 0:
        if row["相关度"] == 1:
            if last_score == 1:
                df_result = df_result.drop(index) 
        last_score = row["相关度"]
        
    if row["相关度"]  > 0.80  and row["相关度"]  != 1:
        df_result = df_result.drop(index)  
    
df_result  = df_result.reset_index(drop=True)
df_result
```

```python
df_result.to_excel('output.xlsx', index=True)
```
