```python
import pandas as pd
```

```python
df = pd.read_excel('real_data.xlsx')
df
```

```python
df_985 = df
for index, row in df_985.iterrows():  
    if row["score"]  < 0.985 :
        df_985 = df_985.drop(index)  
df_985  = df_985.reset_index(drop=True)
df = df_985
```

```python
data_columns = ['dwd_content_id','content_clean']
df = df[data_columns]
df.columns =['id', 'sentence']  
```

```python
for i in range(int(df.shape[0])):
    df.iloc[i,1] = df.iloc[i,1].replace('"',"")
    df.iloc[i,1] = df.iloc[i,1].replace('\n',"")
    df.iloc[i,1] = df.iloc[i,1].replace('\t',"")
    df.iloc[i,1] = df.iloc[i,1].replace('\\',"")
    df.iloc[i,1] = df.iloc[i,1].replace(' '," ")
df   
```

```python
df.to_json('data.json', orient='records')
```

```python
import json

# 打开 JSON 文件
with open('data.json') as f:
    # 读取 JSON 文件内容
    data = json.load(f)
with open('input.json', 'w') as f:
    for obj in data:
        f.write(str(obj).replace("'", '"') + '\n')
```
